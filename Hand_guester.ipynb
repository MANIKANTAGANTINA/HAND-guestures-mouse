```python
import cv2
import mediapipe as mp
import pyautogui
import math
import time

# Initialize MediaPipe
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
capture_hands = mp_hands.Hands()

# Screen size
screen_width, screen_height = pyautogui.size()

# Webcam
camera = cv2.VideoCapture(0)

# Time tracking
last_click_time = 0
click_interval = 0.3  # seconds

while True:
    success, image = camera.read()
    if not success:
        break

    image = cv2.flip(image, 1)
    image_height, image_width, _ = image.shape
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    output_hands = capture_hands.process(rgb_image)
    all_hands = output_hands.multi_hand_landmarks

    x1 = y1 = x2 = y2 = 0

    if all_hands:
        for hand in all_hands:
            mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            landmarks = hand.landmark

            for id, lm in enumerate(landmarks):
                x = int(lm.x * image_width)
                y = int(lm.y * image_height)

                if id == 8:  # Index finger tip
                    x1, y1 = x, y
                    mouse_x = int(screen_width * lm.x)
                    mouse_y = int(screen_height * lm.y)
                    pyautogui.moveTo(mouse_x, mouse_y)
                    cv2.circle(image, (x, y), 10, (0, 255, 255), -1)

                if id == 4:  # Thumb tip
                    x2, y2 = x, y
                    cv2.circle(image, (x, y), 10, (255, 0, 0), -1)

        # Calculate distance between index and thumb
        distance = math.hypot(x2 - x1, y2 - y1)
        print("Distance:", distance)

        # If fingers are close enough, simulate click
        current_time = time.time()
        if distance < 40 and (current_time - last_click_time > click_interval):
            pyautogui.click()
            print("Clicked!")
            last_click_time = current_time

    # Display image
    cv2.imshow("Hand movement video capture", image)
    if cv2.waitKey(1) == 27:
        break

camera.release()
cv2.destroyAllWindows()
import cv2
import mediapipe as mp
import pyautogui
import math
import time

# Initialize MediaPipe
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
capture_hands = mp_hands.Hands()

# Screen size
screen_width, screen_height = pyautogui.size()

# Webcam
camera = cv2.VideoCapture(0)

# Time tracking
last_click_time = 0
click_interval = 0.3  # seconds

while True:
    success, image = camera.read()
    if not success:
        break

    image = cv2.flip(image, 1)
    image_height, image_width, _ = image.shape
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    output_hands = capture_hands.process(rgb_image)
    all_hands = output_hands.multi_hand_landmarks

    x1 = y1 = x2 = y2 = 0

    if all_hands:
        for hand in all_hands:
            mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)
            landmarks = hand.landmark

            for id, lm in enumerate(landmarks):
                x = int(lm.x * image_width)
                y = int(lm.y * image_height)

                if id == 8:  # Index finger tip
                    x1, y1 = x, y
                    mouse_x = int(screen_width * lm.x)
                    mouse_y = int(screen_height * lm.y)
                    pyautogui.moveTo(mouse_x, mouse_y)
                    cv2.circle(image, (x, y), 10, (0, 255, 255), -1)

                if id == 4:  # Thumb tip
                    x2, y2 = x, y
                    cv2.circle(image, (x, y), 10, (255, 0, 0), -1)

        # Calculate distance between index and thumb
        distance = math.hypot(x2 - x1, y2 - y1)
        print("Distance:", distance)

        # If fingers are close enough, simulate click
        current_time = time.time()
        if distance < 40 and (current_time - last_click_time > click_interval):
            pyautogui.click()
            print("Clicked!")
            last_click_time = current_time

    # Display image
    cv2.imshow("Hand movement video capture", image)
    if cv2.waitKey(1) == 27:
        break

camera.release()
cv2.destroyAllWindows()
```
